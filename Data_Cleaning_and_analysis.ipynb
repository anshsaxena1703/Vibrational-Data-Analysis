```python
import pandas as pd
```

```python
df=pd.read_csv("Data 1.csv")
```

```python
df
```

```python
df=df.drop(columns = "timeStamp")
df
```

```python
df=df.drop_duplicates()
df
```

```python
print(df.columns)
```

```python
df
```

```python
df=df.rename(columns={"xtRaccel":"x-acc"})
df
```

```python
df=df.rename(columns={"ytRaccel":"y-acc"})
df
```

```python
df=df.rename(columns={"ztRaccel":"z-acc"})
df
```

```python
df=df.rename(columns={"xtRv":"x-vel"})
df
```

```python
df=df.rename(columns={"ytRv":"y-vel"})
df
```

```python
df=df.rename(columns={"ztRv":"z-vel"})
df
```

```python
df['maxnoise'] = (df['maxnoise'] - df['maxnoise'].mean()) / df['maxnoise'].std() 
df['temperature'] = (df['temperature'] - df['temperature'].mean()) / df['temperature'].std()
df['avgnoise'] = (df['avgnoise'] - df['avgnoise'].mean()) / df['avgnoise'].std()
df['totalAccel'] = (df['totalAccel'] - df['totalAccel'].mean()) / df['totalAccel'].std()
df
```

```python
import matplotlib.pyplot as plt
```

```python
df.plot(x='Time', y='temperature', kind='scatter') 
plt.title('Scatter Plot of temperature vs Time')
plt.xlabel('Time')
plt.ylabel('temperature')
plt.show()
```

```python
df.plot(x='Time', y='avgnoise', kind='scatter') 
plt.title('Scatter Plot of Average Noise vs Time')
plt.xlabel('Time')
plt.ylabel('Average Noise')
plt.show()
```

```python
df.plot(x='Time', y='totalAccel', kind='scatter') 
plt.title('Scatter Plot of Time vs Total Acceleration')
plt.xlabel('Time')
plt.ylabel('Total Acceleration')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Plot the first y-axis column against the x-axis column
plt.plot(df['Time'], df['avgnoise'], label='y1', marker='o')

# Plot the second y-axis column against the x-axis column
plt.plot(df['Time'], df['temperature'], label='y2', marker='s')

plt.plot(df['Time'], df['totalAccel'], label='y3', marker='^')

# Add labels and title
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Plot of Average Noise and Temperature and totalAccel against Time')
plt.legend()

# Display the plot
plt.show()
```
